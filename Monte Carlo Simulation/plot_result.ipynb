```python
def convert_str_int(dataframe):

    # Replace newline characters with spaces and remove the square brackets
    input_str = dataframe.replace('\n', ' ').strip("[]")

    # Split the string into a list of numbers
    numbers_list = input_str.split()

    # Convert the list of strings to a list of integers
    int_list = [int(float(number)) for number in numbers_list]

    return int_list
```
