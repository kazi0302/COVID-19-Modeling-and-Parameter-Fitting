```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
nepal_data = pd.read_csv('nepal_data.csv')
province_1 = pd.read_csv('province1_data.csv')
province_2 = pd.read_csv('province2_data.csv')
province_3 = pd.read_csv('province3_data.csv')
province_4 = pd.read_csv('province4_data.csv')
province_5 = pd.read_csv('province5_data.csv')
province_6 = pd.read_csv('province6_data.csv')
province_7 = pd.read_csv('province7_data.csv')
```

```python
def plot_province(data):
    data.plot(x='Date', y=[data.columns[1], data.columns[2], data.columns[3], data.columns[4], data.columns[5], data.columns[6]], figsize=(10, 5))
    plt.grid()
```

```python
print('Nepal Data')
plot_province(nepal_data)
```

```python
print('Province 1')
plot_province(province_1)
```
